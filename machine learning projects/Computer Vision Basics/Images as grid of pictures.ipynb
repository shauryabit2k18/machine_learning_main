# Import Resources

```python
import numpy as np
import matplotlib.image as mpimg   # for reding in images
import matplotlib.pyplot as plt
import cv2                         # computer vision library
%matplotlib inline
```

# Read In And Display The Images

```python
# Read in the image
image = mpimg.imread('images/waymo_car.jpg')

# Print out the image dimensions
print('image dimentions:' , image.shape)

# Change from RGB to Gray Scale
gray_image = cv2.cvtColor(image , cv2.COLOR_RGB2GRAY)

plt.imshow(gray_image , cmap = 'gray')
```

```python
plt.imshow(image)
```

```python
# Print the paryticular grayscale pixel value
# What is the pixel value at x = 400 and y = 300

x = 400
y = 300

print(gray_image[y , x])
```

```python
# Find the max and min grayscale values in the image
max_val = np.amax(gray_image)
min_val = np.amin(gray_image)

print('max:' , max_val)
print('min:' , min_val)
```

```python
# Create a 5x5 image using just grayscale , numerical values
tiny_image = np.array([[0, 0, 0, 0, 0],
                      [0, 255, 0, 255, 0],
                      [0, 0, 0, 0, 0],
                      [0, 255, 255, 255, 0],
                      [0, 0, 0, 0, 0]])

# To show the pixel grid, use matshow
plt.matshow(tiny_image , cmap = 'gray')
```
